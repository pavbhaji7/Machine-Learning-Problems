```python
# Exercise 7: Hidden Markov Models (HMM) for Sequential Data
# Comprehensive implementation with multiple examples

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from hmmlearn import hmm
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (15, 10)

print("="*70)
print("EXERCISE 7: HIDDEN MARKOV MODELS (HMM)")
print("="*70)

# ============================================================================
# 1. HMM BASICS AND THEORY
# ============================================================================

print("\n1. Hidden Markov Model Components...")
print("""
HMM consists of:
1. States (S): Hidden states that are not directly observable
2. Observations (O): Observable outputs generated by states
3. Transition Matrix (A): Probabilities of moving between states
4. Emission Matrix (B): Probabilities of observations given states
5. Initial State Distribution (π): Starting state probabilities

Three fundamental problems:
1. Evaluation: P(O|λ) - Forward algorithm
2. Decoding: Best state sequence - Viterbi algorithm
3. Learning: Best parameters λ - Baum-Welch algorithm
""")

# ============================================================================
# 2. SIMPLE WEATHER PREDICTION MODEL
# ============================================================================

print("\n" + "="*70)
print("2. WEATHER PREDICTION MODEL")
print("="*70)

print("\n2.1 Creating Simple Weather HMM...")

# Define states and observations
states = ['Sunny', 'Rainy']
observations = ['Walk', 'Shop', 'Clean']

# Create transition matrix
# Rows: current state, Columns: next state
transition_matrix = np.array([
    [0.7, 0.3],  # Sunny -> [Sunny, Rainy]
    [0.4, 0.6]   # Rainy -> [Sunny, Rainy]
])

# Create emission matrix
# Rows: states, Columns: observations
emission_matrix = np.array([
    [0.6, 0.3, 0.1],  # Sunny -> [Walk, Shop, Clean]
    [0.1, 0.4, 0.5]   # Rainy -> [Walk, Shop, Clean]
])

# Initial state distribution
start_probability = np.array([0.6, 0.4])  # [Sunny, Rainy]

print("\nTransition Matrix (State -> State):")
trans_df = pd.DataFrame(transition_matrix,
                        index=states,
                        columns=states)
print(trans_df)

print("\nEmission Matrix (State -> Observation):")
emis_df = pd.DataFrame(emission_matrix,
                       index=states,
                       columns=observations)
print(emis_df)

print("\nInitial State Probabilities:")
init_df = pd.DataFrame({'Probability': start_probability},
                       index=states)
print(init_df)

# Visualize the model
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Transition matrix heatmap
sns.heatmap(transition_matrix, annot=True, fmt='.2f', cmap='YlGnBu',
            xticklabels=states, yticklabels=states, ax=axes[0],
            cbar_kws={'label': 'Probability'})
axes[0].set_title('State Transition Matrix', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Next State')
axes[0].set_ylabel('Current State')

# Emission matrix heatmap
sns.heatmap(emission_matrix, annot=True, fmt='.2f', cmap='YlOrRd',
            xticklabels=observations, yticklabels=states, ax=axes[1],
            cbar_kws={'label': 'Probability'})
axes[1].set_title('Emission Matrix', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Observation')
axes[1].set_ylabel('State')

# Initial probabilities
axes[2].bar(states, start_probability, color=['gold', 'skyblue'],
           edgecolor='black', alpha=0.7)
axes[2].set_ylabel('Probability', fontsize=12)
axes[2].set_title('Initial State Distribution', fontsize=14, fontweight='bold')
axes[2].set_ylim([0, 1])
axes[2].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# 3. IMPLEMENTING HMM WITH hmmlearn
# ============================================================================

print("\n" + "="*70)
print("3. IMPLEMENTING HMM WITH hmmlearn")
print("="*70)

print("\n3.1 Creating and Training Weather HMM...")

# Create observation sequence
# 0: Walk, 1: Shop, 2: Clean
observation_sequence = np.array([0, 0, 1, 2, 2, 1, 0, 0, 2, 1, 0, 0]).reshape(-1, 1)

print(f"\nObservation Sequence: {[observations[i[0]] for i in observation_sequence]}")

# Create HMM model
model = hmm.CategoricalHMM(n_components=2, random_state=42)

# Set parameters
model.startprob_ = start_probability
model.transmat_ = transition_matrix
model.emissionprob_ = emission_matrix

print("\nModel Parameters Set:")
print(f"  • Number of states: {model.n_components}")
print(f"  • Start probabilities: {model.startprob_}")
print(f"  • Transition matrix shape: {model.transmat_.shape}")
print(f"  • Emission matrix shape: {model.emissionprob_.shape}")

# Decode the most likely state sequence (Viterbi algorithm)
logprob, state_sequence = model.decode(observation_sequence, algorithm='viterbi')

print(f"\nLog Probability: {logprob:.4f}")
print(f"\nPredicted State Sequence:")
for i, (obs, state) in enumerate(zip(observation_sequence, state_sequence)):
    print(f"  Day {i+1}: Observed={observations[obs[0]]}, Predicted State={states[state]}")

# Visualize the sequence
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 8))

days = range(1, len(observation_sequence) + 1)

# Observations
obs_numeric = [obs[0] for obs in observation_sequence]
colors_obs = ['green', 'orange', 'red']
ax1.scatter(days, obs_numeric, c=[colors_obs[o] for o in obs_numeric],
           s=200, edgecolors='black', linewidths=2, zorder=3)
ax1.plot(days, obs_numeric, 'k--', alpha=0.3, linewidth=1)
ax1.set_yticks(range(3))
ax1.set_yticklabels(observations)
ax1.set_xlabel('Day', fontsize=12)
ax1.set_ylabel('Observation', fontsize=12)
ax1.set_title('Observable Activity Sequence', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3)

# Hidden states
colors_state = ['gold', 'skyblue']
ax2.scatter(days, state_sequence, c=[colors_state[s] for s in state_sequence],
           s=200, edgecolors='black', linewidths=2, zorder=3)
ax2.plot(days, state_sequence, 'k--', alpha=0.3, linewidth=1)
ax2.set_yticks(range(2))
ax2.set_yticklabels(states)
ax2.set_xlabel('Day', fontsize=12)
ax2.set_ylabel('Hidden State', fontsize=12)
ax2.set_title('Predicted Weather State Sequence (Viterbi)', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate forward probabilities
log_prob_forward = model.score(observation_sequence)
print(f"\n3.2 Forward Algorithm (Evaluation):")
print(f"  • Log probability of observation sequence: {log_prob_forward:.4f}")
print(f"  • Probability: {np.exp(log_prob_forward):.6e}")

# Predict next observation
last_state_prob = model.predict_proba(observation_sequence)[-1]
print(f"\n3.3 State Probabilities after Last Observation:")
for i, (state, prob) in enumerate(zip(states, last_state_prob)):
    print(f"  • {state}: {prob:.4f}")

# Predict next observation probabilities
next_obs_prob = np.dot(last_state_prob, emission_matrix)
print(f"\n3.4 Next Observation Probabilities:")
for obs, prob in zip(observations, next_obs_prob):
    print(f"  • {obs}: {prob:.4f}")

# ============================================================================
# 4. GAUSSIAN HMM FOR CONTINUOUS OBSERVATIONS
# ============================================================================

print("\n" + "="*70)
print("4. GAUSSIAN HMM FOR CONTINUOUS DATA")
print("="*70)

print("\n4.1 Generating Synthetic Stock Price Data...")

# Generate synthetic stock price data with hidden market regimes
np.random.seed(42)

# Define market regimes: Bull, Bear, Sideways
n_samples = 300
regime_durations = [100, 100, 100]

# Generate data
bull_returns = np.random.normal(0.02, 0.01, regime_durations[0])
bear_returns = np.random.normal(-0.015, 0.015, regime_durations[1])
sideways_returns = np.random.normal(0.001, 0.008, regime_durations[2])

returns = np.concatenate([bull_returns, bear_returns, sideways_returns])
true_regimes = np.concatenate([
    np.zeros(regime_durations[0]),
    np.ones(regime_durations[1]),
    np.full(regime_durations[2], 2)
]).astype(int)

# Calculate cumulative prices
initial_price = 100
prices = initial_price * np.exp(np.cumsum(returns))

# Add feature: volatility (rolling standard deviation)
window = 10
volatility = pd.Series(returns).rolling(window=window).std().fillna(0).values

# Combine features
X_stock = np.column_stack([returns, volatility])

print(f"Generated {n_samples} days of stock data")
print(f"Features: Returns and Volatility")
print(f"True regimes: Bull (0), Bear (1), Sideways (2)")

# Visualize the synthetic data
fig, axes = plt.subplots(3, 1, figsize=(15, 12))

# Price
axes[0].plot(prices, linewidth=2, color='darkblue')
axes[0].axvspan(0, 100, alpha=0.2, color='green', label='Bull Market')
axes[0].axvspan(100, 200, alpha=0.2, color='red', label='Bear Market')
axes[0].axvspan(200, 300, alpha=0.2, color='gray', label='Sideways Market')
axes[0].set_ylabel('Price ($)', fontsize=12)
axes[0].set_title('Stock Price with True Market Regimes', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Returns
axes[1].plot(returns, linewidth=1, color='darkgreen', alpha=0.7)
axes[1].axhline(y=0, color='black', linestyle='--', linewidth=1)
axes[1].axvspan(0, 100, alpha=0.2, color='green')
axes[1].axvspan(100, 200, alpha=0.2, color='red')
axes[1].axvspan(200, 300, alpha=0.2, color='gray')
axes[1].set_ylabel('Returns', fontsize=12)
axes[1].set_title('Daily Returns', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Volatility
axes[2].plot(volatility, linewidth=2, color='darkred')
axes[2].axvspan(0, 100, alpha=0.2, color='green')
axes[2].axvspan(100, 200, alpha=0.2, color='red')
axes[2].axvspan(200, 300, alpha=0.2, color='gray')
axes[2].set_xlabel('Day', fontsize=12)
axes[2].set_ylabel('Volatility', fontsize=12)
axes[2].set_title('Rolling Volatility', fontsize=14, fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("\n4.2 Training Gaussian HMM...")

# Add small regularization to features for numerical stability
X_stock_reg = X_stock + np.random.normal(0, 1e-6, X_stock.shape)

# Create and train Gaussian HMM with robust parameters
n_states = 3

# Try different covariance types in order of preference
for cov_type in ["diag", "spherical", "tied"]:
    try:
        model_gaussian = hmm.GaussianHMM(
            n_components=n_states,
            covariance_type=cov_type,
            n_iter=100,
            random_state=42,
            min_covar=1e-3
        )

        model_gaussian.fit(X_stock_reg)
        print(f"Model trained successfully with covariance_type='{cov_type}'")
        break

    except (ValueError, np.linalg.LinAlgError) as e:
        print(f"Warning: Failed with '{cov_type}' covariance")
        if cov_type == "tied":  # Last option
            print("Using simplified 1D data...")
            # Use only returns (1D) as fallback
            X_stock_simple = returns.reshape(-1, 1)
            model_gaussian = hmm.GaussianHMM(
                n_components=n_states,
                covariance_type="diag",
                n_iter=100,
                random_state=42,
                min_covar=1e-2
            )
            model_gaussian.fit(X_stock_simple)
            X_stock_reg = X_stock_simple
            print("Trained with simplified 1D data")
        continue

print(f"\nModel trained with {n_states} hidden states")
print(f"Converged: {model_gaussian.monitor_.converged}")
print(f"Number of iterations: {model_gaussian.monitor_.iter}")

# Predict hidden states
predicted_states = model_gaussian.predict(X_stock_reg)

print("\nLearned Model Parameters:")
print(f"\nTransition Matrix:")
trans_gaussian_df = pd.DataFrame(model_gaussian.transmat_,
                                 columns=[f'State {i}' for i in range(n_states)],
                                 index=[f'State {i}' for i in range(n_states)])
print(trans_gaussian_df)

print(f"\nMeans for each state:")
for i in range(n_states):
    if X_stock_reg.shape[1] == 2:
        print(f"  State {i}: Returns={model_gaussian.means_[i][0]:.5f}, "
              f"Volatility={model_gaussian.means_[i][1]:.5f}")
    else:
        print(f"  State {i}: Returns={model_gaussian.means_[i][0]:.5f}")

# Visualize predictions
fig, axes = plt.subplots(2, 1, figsize=(15, 10))

# True regimes
regime_colors = ['green', 'red', 'gray']
for i in range(n_samples):
    axes[0].axvspan(i, i+1, alpha=0.3, color=regime_colors[true_regimes[i]])
axes[0].plot(prices, linewidth=2, color='darkblue')
axes[0].set_ylabel('Price ($)', fontsize=12)
axes[0].set_title('Stock Price with True Market Regimes', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)

# Predicted regimes
pred_colors = ['purple', 'orange', 'cyan']
for i in range(n_samples):
    axes[1].axvspan(i, i+1, alpha=0.3, color=pred_colors[predicted_states[i]])
axes[1].plot(prices, linewidth=2, color='darkblue')
axes[1].set_xlabel('Day', fontsize=12)
axes[1].set_ylabel('Price ($)', fontsize=12)
axes[1].set_title('Stock Price with HMM Predicted Regimes', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

# Add legend
from matplotlib.patches import Patch
legend_elements_true = [Patch(facecolor=c, alpha=0.3, label=l)
                        for c, l in zip(regime_colors, ['Bull', 'Bear', 'Sideways'])]
legend_elements_pred = [Patch(facecolor=c, alpha=0.3, label=f'State {i}')
                        for i, c in enumerate(pred_colors)]

axes[0].legend(handles=legend_elements_true, loc='upper left')
axes[1].legend(handles=legend_elements_pred, loc='upper left')

plt.tight_layout()
plt.show()

# Calculate accuracy by matching states to regimes
print("\n4.3 State-Regime Mapping Analysis:")
state_regime_counts = pd.crosstab(predicted_states, true_regimes,
                                   rownames=['Predicted State'],
                                   colnames=['True Regime'])
print(state_regime_counts)

# ============================================================================
# 5. SPEECH RECOGNITION SIMULATION
# ============================================================================

print("\n" + "="*70)
print("5. SPEECH RECOGNITION SIMULATION")
print("="*70)

print("\n5.1 Generating Phoneme Sequence Data...")

# Simulate phoneme recognition
# States: phonemes (hidden)
# Observations: acoustic features (observable)

phonemes = ['AH', 'EH', 'IH', 'OH', 'UH']  # 5 phonemes
n_phoneme_states = len(phonemes)

# Generate synthetic acoustic features for phoneme sequence "HELLO"
# H-EH-L-L-OH
np.random.seed(42)

# Define mean acoustic features for each phoneme (2D features)
phoneme_means = {
    'AH': [0.5, 0.3],
    'EH': [0.8, 0.6],
    'IH': [0.3, 0.8],
    'OH': [0.2, 0.2],
    'UH': [0.7, 0.4]
}

# Generate sequence for word "HELLO" -> EH, EH, L, L, OH (simplified)
# Let's use: EH (30 frames), IH (30 frames), OH (30 frames)
n_frames_per_phoneme = 30

sequence_phonemes = ['EH', 'IH', 'OH']
X_speech = []
y_speech_true = []

for phoneme in sequence_phonemes:
    mean = phoneme_means[phoneme]
    # Generate frames with some noise
    frames = np.random.multivariate_normal(mean, [[0.05, 0.01], [0.01, 0.05]],
                                           n_frames_per_phoneme)
    X_speech.append(frames)
    y_speech_true.extend([phonemes.index(phoneme)] * n_frames_per_phoneme)

X_speech = np.vstack(X_speech)
y_speech_true = np.array(y_speech_true)

print(f"Generated {len(X_speech)} frames of acoustic features")
print(f"Phoneme sequence: {' -> '.join(sequence_phonemes)}")
print(f"Features per frame: {X_speech.shape[1]}")

# Visualize acoustic features
plt.figure(figsize=(15, 6))
scatter = plt.scatter(X_speech[:, 0], X_speech[:, 1],
                     c=y_speech_true, cmap='viridis',
                     s=50, alpha=0.6, edgecolors='k')

# Add phoneme labels
for i, phoneme in enumerate(phonemes):
    if phoneme in sequence_phonemes:
        mask = y_speech_true == i
        if np.any(mask):
            center = X_speech[mask].mean(axis=0)
            plt.text(center[0], center[1], phoneme,
                    fontsize=16, fontweight='bold',
                    ha='center', va='center',
                    bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.xlabel('Acoustic Feature 1', fontsize=12)
plt.ylabel('Acoustic Feature 2', fontsize=12)
plt.title('Acoustic Feature Space for Phoneme Sequence', fontsize=14, fontweight='bold')
plt.colorbar(scatter, label='Phoneme Index')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n5.2 Training Phoneme Recognition HMM...")

# Train Gaussian HMM with robust settings
X_speech_reg = X_speech + np.random.normal(0, 1e-6, X_speech.shape)

try:
    model_speech = hmm.GaussianHMM(n_components=3,  # 3 phonemes in our sequence
                                   covariance_type="diag",
                                   n_iter=100,
                                   random_state=42,
                                   min_covar=1e-3)
    model_speech.fit(X_speech_reg)
except (ValueError, np.linalg.LinAlgError):
    print("Trying with spherical covariance...")
    model_speech = hmm.GaussianHMM(n_components=3,
                                   covariance_type="spherical",
                                   n_iter=100,
                                   random_state=42,
                                   min_covar=1e-3)
    model_speech.fit(X_speech_reg)

# Predict phoneme sequence
predicted_phonemes = model_speech.predict(X_speech_reg)

print(f"Model trained successfully")
print(f"Converged: {model_speech.monitor_.converged}")

# Visualize predictions
fig, axes = plt.subplots(3, 1, figsize=(15, 12))

# Feature 1 over time
axes[0].plot(X_speech[:, 0], linewidth=2, color='blue', label='Feature 1')
axes[0].axhline(y=0, color='black', linestyle='--', linewidth=1)
for i in range(len(sequence_phonemes)):
    start = i * n_frames_per_phoneme
    end = (i + 1) * n_frames_per_phoneme
    axes[0].axvspan(start, end, alpha=0.2,
                   color=plt.cm.viridis(i / len(sequence_phonemes)))
axes[0].set_ylabel('Feature 1 Value', fontsize=12)
axes[0].set_title('Acoustic Feature 1 Over Time', fontsize=14, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Feature 2 over time
axes[1].plot(X_speech[:, 1], linewidth=2, color='red', label='Feature 2')
axes[1].axhline(y=0, color='black', linestyle='--', linewidth=1)
for i in range(len(sequence_phonemes)):
    start = i * n_frames_per_phoneme
    end = (i + 1) * n_frames_per_phoneme
    axes[1].axvspan(start, end, alpha=0.2,
                   color=plt.cm.viridis(i / len(sequence_phonemes)))
axes[1].set_ylabel('Feature 2 Value', fontsize=12)
axes[1].set_title('Acoustic Feature 2 Over Time', fontsize=14, fontweight='bold')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

# Predicted states
axes[2].plot(predicted_phonemes, linewidth=2, color='purple',
            marker='o', markersize=4, label='Predicted State')
axes[2].plot(y_speech_true, linewidth=2, color='green', alpha=0.5,
            linestyle='--', label='True Phoneme')
axes[2].set_xlabel('Frame', fontsize=12)
axes[2].set_ylabel('State/Phoneme', fontsize=12)
axes[2].set_title('Predicted vs True Phoneme Sequence', fontsize=14, fontweight='bold')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# ============================================================================
# 6. DNA SEQUENCE ANALYSIS
# ============================================================================

print("\n" + "="*70)
print("6. DNA SEQUENCE ANALYSIS (CpG ISLAND DETECTION)")
print("="*70)

print("\n6.1 Generating Synthetic DNA Sequence...")

# Generate synthetic DNA sequence with CpG islands
# CpG islands are regions with high frequency of CG dinucleotides

np.random.seed(42)

def generate_dna_sequence(length, gc_content):
    """Generate DNA sequence with specified GC content"""
    nucleotides = ['A', 'T', 'G', 'C']

    # Calculate AT content
    at_content = 1 - gc_content

    # Each nucleotide probability
    prob_g = gc_content / 2
    prob_c = gc_content / 2
    prob_a = at_content / 2
    prob_t = at_content / 2

    probs = [prob_a, prob_t, prob_g, prob_c]

    # Normalize to avoid floating point errors
    probs = np.array(probs)
    probs = probs / probs.sum()

    sequence = np.random.choice(nucleotides, size=length, p=probs)
    return ''.join(sequence)

# Generate sequence: non-CpG (200bp) + CpG island (100bp) + non-CpG (200bp)
non_cpg_1 = generate_dna_sequence(200, 0.4)
cpg_island = generate_dna_sequence(100, 0.65)
non_cpg_2 = generate_dna_sequence(200, 0.4)

dna_sequence = non_cpg_1 + cpg_island + non_cpg_2
true_labels = np.concatenate([
    np.zeros(200),
    np.ones(100),
    np.zeros(200)
])

print(f"Generated DNA sequence length: {len(dna_sequence)} bp")
print(f"CpG island location: 200-300 bp")
print(f"\nFirst 50 bp: {dna_sequence[:50]}")
print(f"CpG island (50bp sample): {dna_sequence[200:250]}")

# Calculate GC content in windows
window_size = 20
gc_content = []
for i in range(0, len(dna_sequence) - window_size + 1, 5):
    window = dna_sequence[i:i+window_size]
    gc = (window.count('G') + window.count('C')) / window_size
    gc_content.append(gc)

# Visualize GC content
plt.figure(figsize=(15, 6))
positions = range(0, len(dna_sequence) - window_size + 1, 5)
plt.plot(positions, gc_content, linewidth=2, color='darkblue')
plt.axvspan(200, 300, alpha=0.3, color='red', label='True CpG Island')
plt.axhline(y=0.5, color='gray', linestyle='--', label='50% GC threshold')
plt.xlabel('Position (bp)', fontsize=12)
plt.ylabel('GC Content', fontsize=12)
plt.title('GC Content Along DNA Sequence', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n6.2 Training HMM for CpG Island Detection...")

# Encode nucleotides
nucleotide_map = {'A': 0, 'T': 1, 'G': 2, 'C': 3}
X_dna = np.array([nucleotide_map[n] for n in dna_sequence]).reshape(-1, 1)

# Train HMM with 2 states: non-CpG and CpG
model_dna = hmm.CategoricalHMM(n_components=2, n_iter=1000, random_state=42)
model_dna.fit(X_dna)

# Predict states
predicted_states = model_dna.predict(X_dna)

print(f"Model trained successfully")
print(f"Converged: {model_dna.monitor_.converged}")

print(f"\nLearned Emission Probabilities:")
emission_dna_df = pd.DataFrame(model_dna.emissionprob_,
                               columns=['A', 'T', 'G', 'C'],
                               index=['State 0', 'State 1'])
print(emission_dna_df)

# Determine which state corresponds to CpG island
gc_state_0 = model_dna.emissionprob_[0, 2] + model_dna.emissionprob_[0, 3]
gc_state_1 = model_dna.emissionprob_[1, 2] + model_dna.emissionprob_[1, 3]

cpg_state = 0 if gc_state_0 > gc_state_1 else 1
print(f"\nCpG Island State: State {cpg_state} (GC content: {max(gc_state_0, gc_state_1):.3f})")

# Visualize predictions
fig, axes = plt.subplots(3, 1, figsize=(15, 12))

# True labels
axes[0].plot(true_labels, linewidth=2, color='red')
axes[0].fill_between(range(len(true_labels)), 0, true_labels, alpha=0.3, color='red')
axes[0].set_ylabel('CpG Island', fontsize=12)
axes[0].set_title('True CpG Island Location', fontsize=14, fontweight='bold')
axes[0].set_ylim([-0.1, 1.1])
axes[0].set_yticks([0, 1])
axes[0].set_yticklabels(['Non-CpG', 'CpG'])
axes[0].grid(True, alpha=0.3)

# Predicted labels
predicted_binary = (predicted_states == cpg_state).astype(int)
axes[1].plot(predicted_binary, linewidth=2, color='blue')
axes[1].fill_between(range(len(predicted_binary)), 0, predicted_binary,
                     alpha=0.3, color='blue')
axes[1].set_ylabel('CpG Island', fontsize=12)
axes[1].set_title('HMM Predicted CpG Island Location', fontsize=14, fontweight='bold')
axes[1].set_ylim([-0.1, 1.1])
axes[1].set_yticks([0, 1])
axes[1].set_yticklabels(['Non-CpG', 'CpG'])
axes[1].grid(True, alpha=0.3)

# State probabilities
state_probs = model_dna.predict_proba(X_dna)
axes[2].plot(state_probs[:, cpg_state], linewidth=2, color='purple')
axes[2].fill_between(range(len(state_probs)), 0, state_probs[:, cpg_state],
                     alpha=0.3, color='purple')
axes[2].set_xlabel('Position (bp)', fontsize=12)
axes[2].set_ylabel('CpG Probability', fontsize=12)
axes[2].set_title('Probability of CpG Island State', fontsize=14, fontweight='bold')
axes[2].set_ylim([0, 1])
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Calculate accuracy
accuracy = accuracy_score(true_labels, predicted_binary)
conf_matrix = confusion_matrix(true_labels, predicted_binary)

print(f"\n6.3 Prediction Performance:")
print(f"  • Accuracy: {accuracy:.4f}")
print(f"\nConfusion Matrix:")
print(pd.DataFrame(conf_matrix,
                  columns=['Predicted Non-CpG', 'Predicted CpG'],
                  index=['True Non-CpG', 'True CpG']))

# ============================================================================
# 7. MULTIPLE SEQUENCE COMPARISON
# ============================================================================

print("\n" + "="*70)
print("7. COMPARING DIFFERENT HMM CONFIGURATIONS")
print("="*70)

print("\n7.1 Testing Different Numbers of States...")

# Use the stock data for comparison
n_states_range = range(2, 7)
bic_scores = []
aic_scores = []
log_likelihoods = []

for n in n_states_range:
    try:
        model_temp = hmm.GaussianHMM(n_components=n,
                                     covariance_type="diag",
                                     n_iter=100,
                                     random_state=42,
                                     min_covar=1e-3)
        model_temp.fit(X_stock_reg)

        # Calculate scores
        log_likelihood = model_temp.score(X_stock_reg)
        n_params = n**2 + n * X_stock_reg.shape[1] * 2 + n * X_stock_reg.shape[1]

        aic = 2 * n_params - 2 * log_likelihood * len(X_stock_reg)
        bic = n_params * np.log(len(X_stock_reg)) - 2 * log_likelihood * len(X_stock_reg)

        log_likelihoods.append(log_likelihood)
        aic_scores.append(aic)
        bic_scores.append(bic)

        print(f"\n{n} States:")
        print(f"  • Log-likelihood: {log_likelihood:.2f}")
        print(f"  • AIC: {aic:.2f}")
        print(f"  • BIC: {bic:.2f}")

    except (ValueError, np.linalg.LinAlgError) as e:
        print(f"\n{n} States: Failed to converge")
        log_likelihoods.append(np.nan)
        aic_scores.append(np.nan)
        bic_scores.append(np.nan)

# Plot comparison
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes[0].plot(list(n_states_range), log_likelihoods, 'bo-',
            linewidth=2, markersize=8)
axes[0].set_xlabel('Number of States', fontsize=12)
axes[0].set_ylabel('Log-Likelihood', fontsize=12)
axes[0].set_title('Log-Likelihood vs Number of States', fontsize=14, fontweight='bold')
axes[0].grid(True, alpha=0.3)

axes[1].plot(list(n_states_range), aic_scores, 'ro-',
            linewidth=2, markersize=8)
axes[1].set_xlabel('Number of States', fontsize=12)
axes[1].set_ylabel('AIC', fontsize=12)
axes[1].set_title('AIC vs Number of States (Lower is Better)', fontsize=14, fontweight='bold')
axes[1].grid(True, alpha=0.3)

axes[2].plot(list(n_states_range), bic_scores, 'go-',
            linewidth=2, markersize=8)
axes[2].set_xlabel('Number of States', fontsize=12)
axes[2].set_ylabel('BIC', fontsize=12)
axes[2].set_title('BIC vs Number of States (Lower is Better)', fontsize=14, fontweight='bold')
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Find optimal states (ignoring NaN values)
bic_valid = [b for b in bic_scores if not np.isnan(b)]
if bic_valid:
    optimal_idx = bic_scores.index(min(bic_valid))
    optimal_states = list(n_states_range)[optimal_idx]
    print(f"\n7.2 Optimal Number of States (by BIC): {optimal_states}")
else:
    print(f"\n7.2 Could not determine optimal number of states")

# ============================================================================
# 8. SUMMARY AND APPLICATIONS
# ============================================================================

print("\n" + "="*70)
print("8. SUMMARY AND APPLICATIONS")
print("="*70)

summary_text = """
HIDDEN MARKOV MODEL (HMM) SUMMARY:

1. KEY COMPONENTS:
   • Hidden States: Unobservable system states
   • Observations: Observable outputs from states
   • Transition Probabilities: State-to-state transitions
   • Emission Probabilities: Observation generation from states
   • Initial State Distribution: Starting state probabilities

2. THREE FUNDAMENTAL PROBLEMS:
   a) Evaluation (Forward Algorithm):
      - Calculate probability of observation sequence
      - Time complexity: O(N²T) where N=states, T=time steps

   b) Decoding (Viterbi Algorithm):
      - Find most likely state sequence
      - Optimal path through hidden states

   c) Learning (Baum-Welch Algorithm):
      - Estimate model parameters from data
      - Expectation-Maximization (EM) approach

3. TYPES OF HMMs:
   • Discrete/Categorical HMM: Discrete observations
   • Gaussian HMM: Continuous observations (normal distribution)
   • GMM-HMM: Mixture of Gaussians for emissions
   • Left-to-Right HMM: Sequential processes (speech)

4. PRACTICAL APPLICATIONS:
   • Speech Recognition: Phoneme/word recognition
   • Bioinformatics: Gene finding, CpG islands
   • Finance: Market regime detection
   • Natural Language Processing: POS tagging
   • Activity Recognition: Human behavior analysis
   • Weather Prediction: Climate modeling

5. MODEL SELECTION:
   • Use BIC/AIC for choosing number of states
   • Cross-validation for generalization
   • Domain knowledge for state interpretation
   • Consider computational constraints

6. LIMITATIONS:
   • Markov assumption (current state only depends on previous)
   • Limited memory (first-order)
   • May require many states for complex dynamics
   • Local optima in parameter learning

7. BEST PRACTICES:
   • Normalize/standardize continuous features
   • Try multiple initializations (random_state)
   • Use appropriate covariance type for Gaussian HMM
   • Validate state interpretations with domain knowledge
   • Check convergence of training algorithm
   • Use 'diag' covariance for stability
   • Add regularization (min_covar) to prevent singular matrices
"""

print(summary_text)

print("\n" + "="*70)
print("EXERCISE 7 COMPLETED SUCCESSFULLY!")
print("="*70)
```
